```python
import numpy as np
from matplotlib import pyplot as plt
import sklearn
from skimage import data as imagedata
from skimage import transform
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
from matplotlib.colors import ListedColormap
from sklearn.decomposition import  PCA,FastICA
```

```python
immoon = imagedata.moon()*(1/255.)
impage = transform.resize(imagedata.page(),(512,512))
imcam = imagedata.camera()*(1/255.)
```

```python
combined = np.zeros((512,512,3))
combined[:,:,0] = (0.7*immoon + 0.2* impage+0.1*imcam)
plt.imshow(combined[:,:,0])
plt.gray()
```

```python
combined[:,:,1] = (0.3*immoon + 0.3* impage+0.4*imcam)
combined[:,:,2] = (0.25*immoon + 0.15* impage+0.6*imcam)
```

```python
plt.imshow(combined[:,:,1])
plt.gray()
```

```python
plt.imshow(combined[:,:,2])
plt.gray()
```

```python
transformPCA = PCA(n_components=3)
```

```python
X = combined.reshape((512*512,3))
```

```python
decomposePCA = (transformPCA.fit_transform(X)).reshape((512,512,3))
```

```python
plt.imshow(decomposePCA[:,:,0])
plt.gray()
```

```python
plt.imshow(-decomposePCA[:,:,0])
plt.gray()
```

```python
plt.imshow(decomposePCA[:,:,1])
plt.gray()
```

```python
plt.imshow(decomposePCA[:,:,2])
plt.gray()
```

```python
transformICA = FastICA(n_components=3)
```

```python
decomposeICA = (transformICA.fit_transform(X)).reshape((512,512,3))
```

```python
plt.imshow(decomposeICA[:,:,0])
plt.gray()
```

```python
plt.imshow(decomposeICA[:,:,1])
plt.gray()
```
